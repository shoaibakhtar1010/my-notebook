```python
def coverPoints(A, B):
        length=len(A)
        steps=0
        for i in range(0,length-1):
            x=abs(A[i]-A[i+1])
            y=abs(B[i]-B[i+1])
            steps+=max(x,y)
        return steps
        
```

```python
A = [0, 1, 1]
B = [0, 1, 2]
coverPoints(A,B)
```
