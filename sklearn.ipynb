```python
from sklearn.datasets import make_blobs
```

```python
X,Y=make_blobs(n_samples=1000,n_features=2,centers=3)
```

```python
X
```

```python
Y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
```

```python
km
```

```python
km.fit(X)
```

```python
km.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(3.94108623, -9.05956376)
plt.scatter(9.33069418, -7.93356044)
plt.scatter(-1.44380718, -8.58819055)
```

```python
l1=list(km.labels_)
```
