In order  for heatmap to work properly the table should be in the matrix form .matrix form is the one where index name and column name are categorical that a cell value should match up

```python
tc=tips.corr()
tc
```

```python
ax = sns.heatmap(tc, annot=True) #notation: "annot" not "annote"
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```
